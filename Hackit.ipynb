```python
def categorize_food(food_name):
    veg_keywords = ['vegetarian', 'veg', 'plant-based', 'tofu', 'paneer']
    non_veg_keywords = ['chicken', 'beef', 'fish', 'pork', 'mutton']

    if any(keyword in food_name.lower() for keyword in veg_keywords):
        return 'veg'
    elif any(keyword in food_name.lower() for keyword in non_veg_keywords):
        return 'non-veg'
    else:
        return 'unknown'
df['Food_Type'] = df['Food'].apply(categorize_food)
df.to_csv('/content/drive/MyDrive/your_dataset_no_duplicates.csv', index=False)
```
